```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv('/content/matches.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
mi_ipl_data = df.groupby(['team1', 'team2']).agg({'win_by_runs': 'mean', 'win_by_wickets': 'mean'})
mi_ipl_data
```

```python
reset_indexed_data = mi_ipl_data.reset_index()
reset_indexed_data
```

```python
df.isnull().sum()
```

```python
city_mode = df['city'].mode()[0]
df['city'].fillna(city_mode, inplace=True)

df.drop(columns=['umpire3'], inplace=True)

winner_mode = df['winner'].mode()[0]
player_of_match_mode = df['player_of_match'].mode()[0]
df['winner'].fillna(winner_mode, inplace=True)
df['player_of_match'].fillna(player_of_match_mode, inplace=True)

df['umpire1'].ffill(inplace=True)
df['umpire2'].ffill(inplace=True)
```

```python
df.isnull().sum()
```

```python
matches_played = df.groupby(['team1', 'city']).size().reset_index(name='matches_played')
matches_played
```

```python
aggregations = {
    'win_by_runs': 'mean',
    'win_by_wickets': 'max',
}


agg_result = df.groupby(['team1', 'city']).agg(aggregations).reset_index()
print(agg_result)
```

```python
agg_result = df.groupby(['team1', 'city']).agg(aggregations).reset_index()


agg_result.set_index(['team1', 'city'], inplace=True)


stacked_result = agg_result.stack()


print(stacked_result)
```

```python
columns_to_stack = ['team1', 'team2', 'toss_winner', 'toss_decision']


stacked_data = (df[columns_to_stack]
                .stack()
                .reset_index()
                .rename(columns={'level_2': 'Match', 0: 'Team/Decision'}))


print(stacked_data)
```

```python
unstacked_data = stacked_data.unstack()


print(unstacked_data)
```

```python
id_vars = [ 'season', 'city', 'date' ]


value_vars = ['team1', 'team2', 'toss_winner', 'toss_decision', 'winner']


melted_data = pd.melt(df, id_vars=id_vars,
                      value_vars=value_vars,
                      var_name='variable', value_name='value')


print(melted_data)
```

```python

pivoted_data = melted_data.pivot_table(index=[ 'season', 'city', 'date'],
                                       columns='variable',
                                       values='value',
                                       aggfunc='first').reset_index()


print(pivoted_data)
```

```python
plt.figure(figsize=(10, 6))
sns.distplot(df['season'], hist=True, kde=True, bins=20, color='skyblue', hist_kws={'edgecolor':'black'})
plt.title('Distribution of Seasons')
plt.xlabel('Season')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

distplot function to visualize the distribution of seasons in the DataFrame 'df'. The plot combines a histogram and a kernel density estimation (KDE) curve. Each bar in the histogram represents the frequency of seasons within a specified bin size, while the KDE curve provides a smooth estimate of the probability density function. The plot is titled "Distribution of Seasons", with the x-axis denoting the seasons and the y-axis representing the density.
